## Examples of activation functions

Step
$$ a(z) = \left \{
\begin{eqnarray}
0,\ \text{if}\ z < 0\\
1,\ \text{if}\ z \geq 0 \\
\end{eqnarray}
\right. 
$$

Sigmoid
$$
a(z) = \frac{1}{1+\exp{(-z)}}
$$

Hyperboic tangent
$$
a(z) = \tanh(z)
$$

Rectified linear unit (ReLU)
$$
a(z) = \max(0, z)
$$
