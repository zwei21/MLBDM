```python
import numpy as np
```

```python
def spiral_potential_3b(x, y):
    return -(0.05*x**2 + 0.05*y**2 + 1)*np.exp(-1.02040816326531/(0.05*(x**2 + y**2)*np.sin(0.03*x**2 + 0.03*y**2 + 3*np.arctan2(x, y)) + 1)**2)
# Meshgrid of potential surface
x = np.linspace(-20,20,200)
y = np.linspace(-20,20,200)

[xx,yy] = np.meshgrid(x,y)

# Calculatin Energy for each
z = spiral_potential_3b(xx, yy)

#Plotting the potential Surface
'''fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(x, y, z, cmap=plt.cm.viridis, alpha=0.6)'''
```

```python
import plotly.graph_objects as go

fig = go.Figure(go.Surface(
    contours = {
        "x": {"show": False, "start": 1.5, "end": 2, "size": 0.04, "color":"white"},
        "z": {"show": True, "start": 0.5, "end": 0.8, "size": 5}
    },
    x = x,
    y = y,
    z = z))
fig.update_layout(
        scene = {
            "xaxis": {"nticks": 20},
            "zaxis": {"nticks": 20},
            'camera_eye': {"x": 0, "y": -1, "z": 0.5},
            "aspectratio": {"x": 1, "y": 1, "z": 0.618}
        })
fig.show()
```

```python
def expl_2b_potential(x, y):
    """An Example of 2d, 2-branch spiral potential function generated by symbolic expressions,\
            used for visualisation
    """
    return -(0.05*x**2 + 0.05*y**2 + 1)*np.exp(-1.02040816326531/(0.05*(x**2 + y**2)*\
            np.sin(0.02*x**2 + 0.02*y**2 + 2*np.arctan2(x, y)) + 1)**2)
```

```python
z_2 = expl_2b_potential(xx, yy)
```

```python
fig = go.Figure(go.Surface(
    contours = {
        "x": {"show": False, "start": 1.5, "end": 2, "size": 0.04, "color":"white"},
        "z": {"show": True, "start": 0.5, "end": 0.8, "size": 5}
    },
    x = x,
    y = y,
    z = z_2))
fig.update_layout(
        scene = {
            "xaxis": {"nticks": 20},
            "zaxis": {"nticks": 20},
            'camera_eye': {"x": 0, "y": -1, "z": 0.5},
            "aspectratio": {"x": 1, "y": 1, "z": 0.618}
        })
fig.show()
```
